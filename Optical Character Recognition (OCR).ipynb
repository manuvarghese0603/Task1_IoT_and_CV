### Loading the image

```python
img = cv2.imread('testimg.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

dpi = plt.rcParams['figure.dpi']
height, width, depth = img.shape
figsize = width / float(dpi), height / float(dpi)
plt.figure(figsize=figsize)
plt.imshow(img)
plt.show()
```

### Detecting Character

```python
hi,wi,_ = img.shape
boxes = pytesseract.image_to_boxes(img)
for b in boxes.splitlines():
    #print(b)
    b = b.split(' ')
    #print(b)
    x,y,w,h = int(b[1]),int(b[2]),int(b[3]),int(b[4])
    cv2.rectangle(img,(x,hi-y),(w,hi-h),(255,0,0),1)
    cv2.putText(img,b[0],(x,hi-y-30),cv2.FONT_HERSHEY_COMPLEX,1,(255,0,0),1)

dpi = plt.rcParams['figure.dpi']
height, width, depth = img.shape
figsize = width / float(dpi), height / float(dpi)
plt.figure(figsize=figsize)
plt.imshow(img)
plt.show()

```

### Detecting Words

```python
img = cv2.imread('testimg.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img = cv2.resize(img,(600,600))

hi,wi,_ = img.shape
boxes = pytesseract.image_to_data(img)
#print(boxes)
for x,b in enumerate(boxes.splitlines()):
    # print(b)
    if x!=0:
        b = b.split()
        #print(b)
        if len(b) == 12:
            x,y,w,h = int(b[6]),int(b[7]),int(b[8]),int(b[9])
            cv2.rectangle(img,(x,y),(w+x,h+y),(255,0,0),1)
            cv2.putText(img,b[11],(x,y-10),cv2.FONT_HERSHEY_COMPLEX_SMALL,1,(255,0,0),1)

dpi = plt.rcParams['figure.dpi']
height, width, depth = img.shape
figsize = width / float(dpi), height / float(dpi)
plt.figure(figsize=figsize)
plt.imshow(img)
plt.show()
```
